```python
from pandas import read_csv
from pandas import datetime
from matplotlib import pyplot

def parser(x):
    return datetime.strptime(x, '%Y%m%d')

from pandas import read_csv
series = read_csv('../data/lewis-lake.csv', header=0, parse_dates=[2], index_col=2, squeeze=True, date_parser=parser)
```

```python
series.head()
```

```python
ll = series.ix['1990-01-01':'2016-12-31']
ll = ll.query('STATION_NAME == "LEWIS LAKE DIVIDE WY US" and TOBS > -9999 and TAVG > -9999 and PRCP > -0.9999')
ll.head()
```

```python
import numpy as np
def month(x):
    return '{:%Y-%m-01}'.format(x)

stats = ll.groupby(month).agg({'PRCP' : np.sum, 'TAVG': np.mean, 'TMAX': np.max, 'TMIN': np.min})
stats.to_csv('../data/yellowstone-wr.csv', index_label='month')
```
