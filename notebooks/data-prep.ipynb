```python
from pandas import read_csv
from pandas import datetime
from matplotlib import pyplot

def parser(x):
    return datetime.strptime(x, '%Y%m%d')

from pandas import read_csv
series = read_csv('../data/lewis-lake.csv', header=0, parse_dates=[2], index_col=2, squeeze=True, date_parser=parser)
```

```python
series.head()
```

```python
ll = series.ix['1990-01-01':'2016-12-31']
ll = ll.query('STATION_NAME == "LEWIS LAKE DIVIDE WY US" and TOBS > -9999 and TAVG > -9999 and PRCP > -0.9999')
ll.head()
```

```python
import numpy as np
def month(x):
    return '{:%Y-%m-01}'.format(x)

weather = ll.groupby(month).agg({'PRCP' : np.sum, 'TAVG': np.mean, 'TMAX': np.max, 'TMIN': np.min})
weather.to_csv('../data/yellowstone-wr.csv', index_label='month')
```

```python
weather
```

```python
def us_ur_parser(y,m):
    x = '{}-{}-01'.format(y,m.replace('M',''))
    return datetime.strptime(x, '%Y-%m-%d')
urate = read_csv('../data/us-unemployment-rate.csv', header=0, parse_dates={'month' : [1, 2]}, index_col=[0], squeeze=True, date_parser=us_ur_parser)
```

```python
urate = urate.ix['1990-01-01':'2016-12-31'].rename(index=str, columns={"Value": "unemployment"})
urate
```

```python
def ysv_parser(x):
    return datetime.strptime(x, '%Y-%m-%d')
vistors = read_csv('../data/yellowstone-visitors.csv', header=0, parse_dates=[0], index_col=0, squeeze=True, date_parser=ysv_parser)
```

```python
vistors = vistors.ix['1990-01-01':'2016-12-31'].rename(index=str, columns={"Value": "unemplyment"})
```

```python
from pandas import concat
combined_features = concat([vistors, urate, weather], axis=1).drop('Series id', axis=1)
combined_features
```

```python
combined_features.to_csv('../data/yellowstone-visitors-ur-weather.csv', float_format='%.1f', index_label='month')
```
